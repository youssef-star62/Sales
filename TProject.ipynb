```python
# Load the CSV file into a pandas DataFrame
df = pd.read_csv('Sales.csv')

# Convert the 'Order Date' to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')

# Extract month from 'Order Date' and insert it after 'Order Date'
df.insert(df.columns.get_loc('Order Date') + 1, 'Month', df['Order Date'].dt.month)

# Safely extract the city from 'Purchase Address' and insert it after 'Purchase Address'
def extract_city(address):
    if isinstance(address, str):
        parts = address.split(',')
        if len(parts) > 1:
            return parts[1].strip()
    return None

df.insert(df.columns.get_loc('Purchase Address') + 1, 'City', df['Purchase Address'].apply(extract_city))

# Convert 'Quantity Ordered' and 'Price Each' to numeric
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'], errors='coerce')
df['Price Each'] = pd.to_numeric(df['Price Each'], errors='coerce')

# Calculate total amount (Quantity Ordered * Price Each) and insert it after 'Price Each'
df.insert(df.columns.get_loc('Price Each') + 1, 'Amount', df['Quantity Ordered'] * df['Price Each'])

# Display the updated DataFrame with the columns near the original ones
print(df.head())
```

```python
df.to_csv('Sales.csv', index=False)
```

```python
# df['Year'] = df['Order Date'].dt.year
df['Quarter'] = df['Order Date'].dt.to_period('Q')  # Quarters
```

```python
bins = [0, 50, 100, 500, 1000, float('inf')]
labels = ['Low', 'Medium', 'High', 'Very High', 'Extremely High']
df['Price Category'] = pd.cut(df['Price Each'], bins=bins, labels=labels)
```

```python
df.to_csv('Sales.csv', index=False)
```

```python
df['City'] = df['City'].str.title()  # Capitalize each word in city names
```

```python
df['Price Each'] = df['Price Each'].round(2)
df['Amount'] = df['Amount'].round(2)
```

```python
df['City'] = df['City'].astype('category')
```

```python
df.to_csv('Sales.csv', index=False)
```

```python
import matplotlib.pyplot as plt






# Sales trend over time
sales_trend = df.groupby([ 'Month'])['Amount'].sum().reset_index()

# Plot sales trends over time
plt.figure(figsize=(10,6))
plt.plot(sales_trend['Month'], sales_trend['Amount'], marker='o')
plt.title('Sales Trend Over Time')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

```python
# Group by city and calculate total sales
city_sales = df.groupby('City')['Amount'].sum().reset_index()
city_sales = city_sales.sort_values(by='Amount', ascending=False)

# Display the top cities by total sales
print(city_sales)

# Bar plot for city sales distribution
plt.figure(figsize=(12,6))
plt.bar(city_sales['City'], city_sales['Amount'], color='skyblue')
plt.xticks(rotation=90)
plt.title('Total Sales by City')
plt.xlabel('City')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Sales trend by product category
product_sales_trend = df.groupby('Product')['Amount'].sum().reset_index().sort_values(by='Amount', ascending=False)

# Bar plot for product category sales
plt.figure(figsize=(12,6))
product_sales_trend.plot(kind='bar', x='Product', y='Amount', legend=False)
plt.title('Sales by Product Category')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Question 1: What was the best months for sales ? How much was earned that month ?
monthly_sales = df.groupby('Month')['Amount'].sum().reset_index()

# Find the best month by locating the maximum sales value
best_month = monthly_sales.loc[monthly_sales['Amount'].idxmax()]

# Display the best month and how much was earned
print(f"The best month for sales was Month: {int(best_month['Month'])}, with total earnings of ${best_month['Amount']:.2f}.")

# Plot sales by month
plt.figure(figsize=(10,6))
plt.bar(monthly_sales['Month'], monthly_sales['Amount'], color='skyblue')
plt.title('Total Sales by Month')
plt.xlabel('Month')
plt.ylabel('Total Sales ($)')
plt.xticks(monthly_sales['Month'])
plt.show()
```

```python
# Question 2: What city had the highest number of sales ?
city_sales = df.groupby('City')['Quantity Ordered'].sum().reset_index()

# Find the city with the highest number of sales
best_city = city_sales.loc[city_sales['Quantity Ordered'].idxmax()]

# Display the city with the highest number of sales
print(f"The city with the highest number of sales is {best_city['City']} with {int(best_city['Quantity Ordered'])} items sold.")

plt.figure(figsize=(12,6))
plt.bar(city_sales['City'], city_sales['Quantity Ordered'], color='skyblue')
plt.xticks(rotation=90)
plt.title('Number of Sales by City')
plt.xlabel('City')
plt.ylabel('Number of Sales')
plt.show()
```

```python
# Question 3: What time should we display advertisements to maximize likelihood of customer's buying product?
# Extract the hour from 'Order Date'
df['Hour'] = df['Order Date'].dt.hour

# Group by hour and count the number of orders (or sum Quantity Ordered)
hourly_sales = df.groupby('Hour')['Quantity Ordered'].count().reset_index()

# Plot sales by hour
plt.figure(figsize=(10,6))
plt.plot(hourly_sales['Hour'], hourly_sales['Quantity Ordered'], marker='o', color='skyblue')
plt.xticks(hourly_sales['Hour'])  # Show each hour on the x-axis
plt.title('Number of Sales by Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Sales')
plt.grid(True)
plt.show()

# Display the hour with the highest number of sales
peak_hour = hourly_sales.loc[hourly_sales['Quantity Ordered'].idxmax()]
print(f"The peak hour for sales is {int(peak_hour['Hour'])}:00 with {int(peak_hour['Quantity Ordered'])} sales.")

df['Day of Week'] = df['Order Date'].dt.day_name()
weekday_sales = df.groupby('Day of Week')['Quantity Ordered'].count().reset_index()
print(weekday_sales)
```

```python
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekday_sales['Day of Week'] = pd.Categorical(weekday_sales['Day of Week'], categories=days_order, ordered=True)
weekday_sales = weekday_sales.sort_values('Day of Week')

# Plot the sales by day of the week using a line chart
plt.figure(figsize=(10,6))
plt.plot(weekday_sales['Day of Week'], weekday_sales['Quantity Ordered'], marker='o', color='blue', linestyle='-', linewidth=2)
plt.title('Number of Sales by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Sales')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Question 5: What Product sold the most or least ? Why do you think it sold the most/least ?
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'], errors='coerce')

# Group by Product and sum the quantity ordered
product_sales = df.groupby('Product')['Quantity Ordered'].sum().reset_index()

# Identify the product that sold the most and least
most_sold_product = product_sales.loc[product_sales['Quantity Ordered'].idxmax()]
least_sold_product = product_sales.loc[product_sales['Quantity Ordered'].idxmin()]

# Display the results
print(f"Most sold product: {most_sold_product['Product']} with {most_sold_product['Quantity Ordered']} units sold.")
print(f"Least sold product: {least_sold_product['Product']} with {least_sold_product['Quantity Ordered']} units sold.")

# Plot product sales
plt.figure(figsize=(12,6))
plt.bar(product_sales['Product'], product_sales['Quantity Ordered'], color='skyblue')
plt.xticks(rotation=90)
plt.title('Total Sales by Product')
plt.xlabel('Product')
plt.ylabel('Quantity Ordered')
plt.show()

# ost sold product: AAA Batteries (4-pack) with 30380.0 units sold.Price for each =3.00 $
# Least sold product: LG Dryer with 634.0 units sold.Price for each = 600 $
```

```python
import pandas as pd

# Load the CSV file
data = pd.read_csv('Sales.csv')  # Replace with your actual CSV file path

# Create a dictionary to map cities to their latitude and longitude
city_coordinates = {
    'New York City': (40.7128, -74.0060),
    'San Francisco': (37.7749, -122.4194),
    'Dallas': (32.7767, -96.7970),
    'Atlanta': (33.7490, -84.3880),
    'Boston': (42.3601, -71.0589),
    'Seattle': (47.6062, -122.3321),
    'Los Angeles': (34.0522, -118.2437),
    'Portland': (45.5152, -122.6784),
    'Austin': (30.2672, -97.7431)
}

# Define a function to extract latitude and longitude from the dictionary
def get_latitude(city):
    return city_coordinates.get(city, (None, None))[0]

def get_longitude(city):
    return city_coordinates.get(city, (None, None))[1]

# Create new columns 'Latitude' and 'Longitude' based on the city column
data['Latitude'] = data['City'].apply(get_latitude)
data['Longitude'] = data['City'].apply(get_longitude)

# Save the updated DataFrame to a new CSV file
data.to_csv('Sales.csv', index=False)

# Display the updated DataFrame
print(data)
```
